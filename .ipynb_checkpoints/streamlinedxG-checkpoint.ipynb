```python
from adjustText import adjust_text
import pandas as pd
import numpy as np
import json
import matplotlib.pyplot as plt
from mplsoccer import Pitch, Sbopen, VerticalPitch
import matplotlib.patches as mpatches
from matplotlib.lines import Line2D
import statsmodels.api as sm
import statsmodels.formula.api as smf
import os
import mplcursors
from pathlib import Path
import warnings
pd.options.mode.chained_assignment = None
warnings.filterwarnings('ignore')
```

## Data set up

```python
file_name = r"C:\Users\lukec\AucklandUnitedShots2023-24.xlsx"
data = pd.read_excel(file_name, sheet_name='Total')
df_shots = pd.DataFrame(data)
df_shots['Goal'] = df_shots['Goal'].astype(object)
df_shots['angle'] = np.where(np.arctan(7.32 * (120-df_shots["x"])/ ((120-df_shots["x"])**2 + df_shots["c"]**2 - (7.32/2)**2)) > 0, np.arctan(7.32 * (120-df_shots["x"]) /((120-df_shots["x"])**2 + df_shots["c"]**2 - (7.32/2)**2)), np.arctan(7.32 * (120-df_shots["x"]) /((120-df_shots["x"])**2 + df_shots["c"]**2 - (7.32/2)**2)) + np.pi)
model_variables = ['angle', 'distance']
model = ''
for v in model_variables[:-1]:
    model = model + v + ' + '
model = model + model_variables[-1]

test_model = smf.glm(formula="Goal ~ " + model, data = df_shots,
                     family = sm.families.Binomial()).fit()
#print(test_model.summary())
b = test_model.params
def calculate_xG(sh):
    bsum = b[0]
    for i,v in enumerate(model_variables):
        bsum=bsum+b[i+1]*sh[v]
    xG = 1/(1+np.exp(bsum))
    return xG

xG = df_shots.apply(calculate_xG, axis=1)
df_shots = df_shots.assign(xG=xG)
pgoal_2d = np.zeros((80,80))
for x in range(80):
    for y in range(80):
        sh=dict()
        a= np.arctan(7.32*x/(x**2 + (y-40)**2 - (7.32/2)**2))
        if a<0:
            a=np.pi+a
        sh['angle'] = a
        sh['distance'] = np.sqrt(x**2 +abs(y-40)**2)
        
        pgoal_2d[x,y] = calculate_xG(sh)


aufc_Dict = {1:"Mitchell", 2: "Green", 3:"McConnel", 4:"Macintosh", 5:"Elliot", 6:"Sheldon", 7:"Knott", 8:"Canham", 9:"Nathan",
             10:"Roche", 11:"Wasi", 12:"Granger", 13:"Collins", 14:"Cook", 15:"Vosper", 16:"Harashima", 17:"Brill", 18:"Pivac", 19:"O'Brien", 20:"Vosper", 0:"Unknown"}
def game_setup(team_name):
    game = df_shots.loc[df_shots["Against"] == team_name]
    goal = game.loc[game["Outcome"] == "goal"]
    off = game.loc[game["Outcome"] == "offGoal"]
    on = game.loc[game["Outcome"]=="onGoal"]
    blocked = game.loc[game["Outcome"]=="blocked"]

    pitch = VerticalPitch(pad_bottom=0.5,
                        half=True,
                        goal_type='box',
                        goal_alpha=0.8)
    fig, ax = pitch.draw(figsize=(12,10))

    sc1 = pitch.scatter(off.x, off.y,
                        s=(off.xG * 1500)+100,
                        edgecolors='#13293f66',
                        c='None',
                        marker ='o',
                        ax=ax)

    sc2 = pitch.scatter(goal.x, goal.y,
                        s=(goal.xG * 1500)+100,
                        c = '#ff6300',
                        marker = 'o',
                        ax=ax)

    sc3 = pitch.scatter(on.x, on.y,
                        s=(on.xG * 1500)+100,
                        c= '#13293fa6',
                        marker ='o',
                        ax=ax)

    sc3 = pitch.scatter(blocked.x, blocked.y,
                        s=(blocked.xG * 1500)+100,
                        c= '#13293f66',
                        marker ='o',
                        ax=ax)
    texts = []
    for i in range(len(game)):
        game['Player'] = game['Player'].fillna(0)
        text_data = "{}, {:.2f}, {}".format(aufc_Dict[game['Player'].iloc[i]], game['xG'].iloc[i], game['Half'].iloc[i])
        text = ax.text(game['y'].iloc[i],
                    game['x'].iloc[i],
                    text_data,
                    color= '#ff6300'if game['Outcome'].iloc[i] == "goal" else '#13293f', fontsize=10)
        texts.append(text)

    adjust_text(texts, arrowprops=dict(arrowstyle='->', color='black'), expand=(2.5, 2.5))
    goal_patch = Line2D([],[], color = 'white', markerfacecolor= '#ff6300', marker = 'o', label = 'Goal', markersize=15)
    on_patch = Line2D([],[], color = 'white', markerfacecolor= '#13293fa6', marker = 'o', label = 'On target', markersize=15)
    off_patch = Line2D([],[], color = 'white', markerfacecolor= 'None',markeredgecolor='#13293f66', marker = 'o', label = 'Off target', markersize=15)
    blocked_patch = Line2D([],[], color = 'white',  markerfacecolor= '#13293f66', marker = 'o', label = 'Blocked', markersize=15)
    ax.legend(handles=[goal_patch, on_patch, off_patch, blocked_patch], loc = 'lower right')
    shot_text = ("Shots: {}\n"
                 "Goals: {}\n"
                 "On target: {}\n"
                 "Off target: {}\n"
                 "Blocked: {}\n"
                 "Expected goals: {:.2f}\n"
                 "Difference: {:.2f}").format(game.shape[0], goal.shape[0], (on.shape[0] + goal.shape[0]), off.shape[0], blocked.shape[0],
                                              round(game['xG'].sum(),2), round(goal.shape[0]-game['xG'].sum(),2))
    ax.text(2, 62, shot_text, bbox={'facecolor':'white', 'pad':10})
```
